```python
X2_train=pd.read_excel(r'F:\gradiation project\camera work\X_train.xlsx')
y2_train=pd.read_excel(r'F:\gradiation project\camera work\y_train.xlsx')
X2_test=pd.read_excel(r'F:\gradiation project\camera work\X_test.xlsx')
y2_test=pd.read_excel(r'F:\gradiation project\camera work\y_test.xlsx')
y2_train
```

```python
y2_test=y2_test.values.ravel()
y2_train=y2_train.values.ravel()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=11) #139
rfc.fit(X2_train,y2_train)
y_pred=rfc.predict(X2_test)
print("test accuracy =",accuracy_score(y2_test,y_pred)*100)
```

```python
from sklearn.metrics import plot_confusion_matrix
class_names=["0","1"]
titles_options = [("Confusion matrix, without normalization", None),
                  ("Normalized confusion matrix", 'true')]
for title, normalize in titles_options:
    disp = plot_confusion_matrix(rfc, X2_test, y2_test,
                                 display_labels=class_names,
                                 cmap=plt.cm.Blues,
                                 normalize=normalize)
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```

```python
aug.drop("Var22",axis=1,inplace=True)
```

```python
X_train=pd.read_excel(r'F:\gradiation project\camera work\X_train_aug.xlsx')
y_train=pd.read_excel(r'F:\gradiation project\camera work\y_train_aug.xlsx')
X_test=pd.read_excel(r'F:\gradiation project\camera work\X_test_aug.xlsx')
y_test=pd.read_excel(r'F:\gradiation project\camera work\y_test_aug.xlsx')
```

```python
y_test=y_test.values.ravel()
y_train=y_train.values.ravel()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc2=RandomForestClassifier(random_state=56) #138
rfc2.fit(X_train,y_train)
y_pred=rfc2.predict(X_test)
print("test accuracy =",accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import plot_confusion_matrix
class_names=["0","1"]
titles_options = [("Confusion matrix, without normalization", None),
                  ("Normalized confusion matrix", 'true')]
for title, normalize in titles_options:
    disp = plot_confusion_matrix(rfc2, X_test, y_test,
                                 display_labels=class_names,
                                 cmap=plt.cm.Blues,
                                 normalize=normalize)
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)

print("KNN test accuracy=",accuracy_score(y_test,knn.predict(X_test))*100)
```

```python
y_pred=lgb.predict(X_test)
```
